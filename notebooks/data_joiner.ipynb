# Concatenating Data

```python
gl = pd.read_csv('../raw_data/gleague.csv')
it = pd.read_csv('../raw_data/international.csv')
nba = pd.read_csv('../raw_data/nba.csv')
ncaa = pd.read_csv('../raw_data/ncaa.csv')
```

```python
gl.shape, it.shape, nba.shape, ncaa.shape
```

```python
gl['Competition'] = 'G-League'
gl.head()
```

```python
it['Competition'] = 'International'
it.head()
```

```python
nba.head()
```

```python
ncaa['Competition'] = 'NCAA'
ncaa.head()
```

```python
df = pd.concat([gl, it, nba, ncaa])
df.shape
```

```python
df.head()
```

```python
df.replace('-', None)
df.replace('NaN', None)
```
